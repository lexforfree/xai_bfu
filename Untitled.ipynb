```python
import   as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
N = 5_000
X = np.random.rand(N, 2)  # [x, y] в [0, 1]

x = X[:, 0]
y = X[:, 1]

# метка: внутри четверти единичного круга
labels = (x**2 + y**2 <= 1).astype(int)

labels[:5], X[:5]
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(x[labels == 1], y[labels == 1], s=5, alpha=0.5, label="in")
plt.scatter(x[labels == 0], y[labels == 0], s=5, alpha=0.5, label="out")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quarter of unit circle")
plt.show()
```

```python
X_inside = X[labels == 1]
X_outside = X[labels == 0]

P_inside = len(X_inside) / len(X)
P_outside = len(X_outside) / len(X)

P_inside, P_outside
```

```python
# inside
mu_in = X_inside.mean(axis=0)
var_in = X_inside.var(axis=0)

# outside
mu_out = X_outside.mean(axis=0)
var_out = X_outside.var(axis=0)

mu_in, var_in, mu_out, var_out
```

```python
bins = 10
edges = np.linspace(0, 1, bins + 1)
edges
```

```python
x_bins = np.digitize(x, edges) - 1
y_bins = np.digitize(y, edges) - 1

# защита от выхода за границы
x_bins = np.clip(x_bins, 0, bins - 1)
y_bins = np.clip(y_bins, 0, bins - 1)
(x_bins, y_bins)
```

```python
# P(class)
classes = [0, 1]  # 0 = outside, 1 = inside
N_total = len(labels)

P_class = {}
for c in classes:
    P_class[c] = np.sum(labels == c) / N_total

P_class, classes
```

```python
# создаём массивы для P(x_bin|class) и P(y_bin|class)
# размер: bins x num_classes
P_x_given_class = np.zeros((bins, 2))
P_y_given_class = np.zeros((bins, 2))

for c in classes:
    # все точки этого класса
    mask = (labels == c)
    
    # x
    for b in range(bins):
        P_x_given_class[b, c] = np.sum(x_bins[mask] == b) / np.sum(mask)
    
    # y
    for b in range(bins):
        P_y_given_class[b, c] = np.sum(y_bins[mask] == b) / np.sum(mask)

P_x_given_class, P_y_given_class
```

```python
def predict_log_posterior(x_val, y_val):
    # бин для точки
    x_b = np.digitize([x_val], edges)[0] - 1
    y_b = np.digitize([y_val], edges)[0] - 1
    x_b = np.clip(x_b, 0, bins - 1)
    y_b = np.clip(y_b, 0, bins - 1)
    
    log_post = {}
    for c in classes:
        # логарифм, чтобы умножение частот не уходило в 0
        log_post[c] = np.log(P_class[c]) + np.log(P_x_given_class[x_b, c] + 1e-9) + np.log(P_y_given_class[y_b, c] + 1e-9)
    
    return log_post
```

```python
def predict_class(x_val, y_val):
    log_post = predict_log_posterior(x_val, y_val)
    return max(log_post, key=log_post.get)
```

```python
test_point = (0.9, 0.9)
pred_class = predict_class(*test_point)
pred_class
```

```python
# создаём сетку точек
xx, yy = np.meshgrid(np.linspace(0,1,200), np.linspace(0,1,200))
grid_points = np.c_[xx.ravel(), yy.ravel()]

# предсказываем класс для каждой точки сетки
pred_grid = np.array([predict_class(xv, yv) for xv, yv in grid_points])
pred_grid = pred_grid.reshape(xx.shape)

# рисуем
plt.figure(figsize=(6,6))
plt.contourf(xx, yy, pred_grid, alpha=0.3, levels=[-0.5,0.5,1.5], colors=['red','green'])

# точки исходного датасета
plt.scatter(x[labels==1], y[labels==1], s=10, alpha=0.5, label='inside')
plt.scatter(x[labels==0], y[labels==0], s=10, alpha=0.5, label='outside')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Naive Bayes (discrete) decision boundary")
plt.legend()
plt.axis("equal")
plt.show()
```
