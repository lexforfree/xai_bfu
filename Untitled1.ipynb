```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5000
X = np.random.rand(N, 2)  # x и y в [0,1]
x = X[:, 0]
y = X[:, 1]
labels = (x**2 + y**2 <= 1).astype(int)
```

```python
x_bins = np.round(x, 1)
y_bins = np.round(y, 1)
```

```python
classes = [0, 1]  # 0 = out, 1 = in
N_total = len(labels)

# Приорные вероятности
P_class = {c: np.sum(labels==c)/N_total for c in classes}
P_class
```

```python
unique_vals = np.arange(0, 1.01, 0.1)  # 0., 0.1, ..., 1.

P_x_given_class = {c: {} for c in classes}
P_y_given_class = {c: {} for c in classes}

for c in classes:
    mask = (labels == c)
    for val in unique_vals:
        P_x_given_class[c][val] = np.sum(x_bins[mask]==val)/np.sum(mask)
        P_y_given_class[c][val] = np.sum(y_bins[mask]==val)/np.sum(mask)
# unique_vals
P_x_given_class, P_y_given_class
```

```python
epsilon = 1e-9  #  log(0)

def predict_log_posterior(x_val, y_val):
    x_b = round(x_val, 1)
    y_b = round(y_val, 1)
    
    log_post = {}
    for c in classes:
        # берем вероятность, если 0 → подставляем epsilon
        px = P_x_given_class[c].get(x_b, 0)  # если ключ не найден, 0
        py = P_y_given_class[c].get(y_b, 0)
        
        px = max(px, epsilon)
        py = max(py, epsilon)
        
        log_post[c] = np.log(P_class[c]) + np.log(px) + np.log(py)
    return log_post
```

```python
def predict_class(x_val, y_val):
    log_post = predict_log_posterior(x_val, y_val)
    return max(log_post, key=log_post.get)
```

```python
xx, yy = np.meshgrid(np.linspace(0,1,200), np.linspace(0,1,200))
grid_points = np.c_[xx.ravel(), yy.ravel()]
pred_grid = np.array([predict_class(xv, yv) for xv, yv in grid_points])
pred_grid = pred_grid.reshape(xx.shape)

plt.figure(figsize=(10,10))
plt.contourf(xx, yy, pred_grid, alpha=0.24, levels=[-0.5,0.5,1.5], colors=['red','green'])
plt.scatter(x[labels==1], y[labels==1], s=10, alpha=0.5, label='inside')
plt.scatter(x[labels==0], y[labels==0], s=10, alpha=0.5, label='outside')
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.axis("equal")
plt.show()
```
